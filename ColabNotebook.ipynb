```python

import numpy as np, csv, math, os

# Load your waypoints
data = []
with open('levine_waypoints.csv') as f:
    reader = csv.reader(f)
    next(reader)  # skip header
    for row in reader:
        data.append(list(map(float, row)))
wp = np.array(data)  # [x, y, yaw]

# Compute cumulative arc-length
diffs = np.diff(wp[:, :2], axis=0)
ds = np.hypot(diffs[:, 0], diffs[:, 1])
# We use 0.1 as the start but keep the total count at 281 to avoid shape mismatches
s = np.concatenate([[0.1], 0.1 + np.cumsum(ds)])

x, y, yaw = wp[:, 0], wp[:, 1], wp[:, 2]
vx = np.full(len(x), 3.0)   # placeholder speed
ax = np.zeros(len(x))
kappa = np.zeros(len(x))

lines = [
    '# generated from levine_waypoints.csv',
    '# placeholder',
    '# s_m; x_m; y_m; psi_rad; kappa_radpm; vx_mps; ax_mps2',
]
for i in range(len(x)):
    lines.append(f"{s[i]:.6f}; {x[i]:.6f}; {y[i]:.6f}; {yaw[i]:.6f}; {kappa[i]:.6f}; {vx[i]:.6f}; {ax[i]:.6f}")
# Close the loop: repeat first point with total arc-length as s value
# Required by scipy CubicSpline bc_type='periodic'
s_close = float(s[-1]) + float(np.hypot(x[-1] - x[0], y[-1] - y[0]))
lines.append(f"{s_close:.6f}; {x[0]:.6f}; {y[0]:.6f}; {yaw[0]:.6f}; {kappa[0]:.6f}; {vx[0]:.6f}; {ax[0]:.6f}")

os.makedirs('/content/f1tenth_gym/maps/levine', exist_ok=True)
os.makedirs('/content/f1tenth_gym/f1tenth_gym/maps/levine', exist_ok=True)
raceline_content = '\n'.join(lines)
with open('/content/f1tenth_gym/maps/levine/levine_raceline.csv', 'w') as f:
    f.write(raceline_content)
with open('/content/f1tenth_gym/f1tenth_gym/maps/levine/levine_raceline.csv', 'w') as f:
    f.write(raceline_content)
print(f"Done. Generated {len(x)+1} points (loop closed) with s[0]=0.1")
```

```python
import numpy as np
import gymnasium as gym
from f1tenth_gym.envs.env_config import EnvConfig, ObservationConfig
from f1tenth_gym.envs.observation import ObservationType

# Setup environment to inspect occupancy logic
cfg = EnvConfig(
    map_name="levine",
    observation_config=ObservationConfig(type=ObservationType.KINEMATIC_STATE)
)
env = gym.make("f1tenth_gym:f1tenth-v0", config=cfg)
track = env.unwrapped.track

def is_pixel_colliding(x, y):
    # Convert world (x,y) to map pixel indices
    # origin is [x, y, yaw]
    res = track.spec.resolution
    origin_x = track.spec.origin[0]
    origin_y = track.spec.origin[1]

    # Coordinate transformation similar to gym internal logic
    # Note: Occupancy map is usually [H, W]
    col = int((x - origin_x) / res)
    row = int((y - origin_y) / res)

    if 0 <= row < track.occupancy_map.shape[0] and 0 <= col < track.occupancy_map.shape[1]:
        # 0 is usually free, 1 or 255 is occupied depending on map loading
        # In f1tenth_gym, occupied_thresh defines the boundary
        val = track.occupancy_map[row, col]
        return val > 0
    return True # Out of bounds is collision

print(f"--- Map Info ---")
print(f"Resolution: {track.spec.resolution}")
print(f"Origin: {track.spec.origin}")
print(f"Map shape: {track.occupancy_map.shape}")

# Check collision at start point (0,0)
x, y = 0.0, 0.0
colliding = is_pixel_colliding(x, y)

print(f"\n--- Collision Check ---")
print(f"Position: ({x}, {y})")
print(f"Is in collision (occupied > 0): {colliding}")

# Sample a small area around the start
sample_range = np.linspace(-0.5, 0.5, 5)
print("\n--- Collision Map (0.25m grid around origin) ---")
for sy in reversed(sample_range):
    row_str = []
    for sx in sample_range:
        c = is_pixel_colliding(sx, sy)
        row_str.append("X" if c else ".")
    print(f"y={sy: .2f}  " + " ".join(row_str))
print("          " + " ".join([f"{sx:.2f}" for sx in sample_range]))

env.close()
```

```python
import time
import gymnasium as gym
import gymnasium.wrappers
import numpy as np
import f1tenth_gym

from f1tenth_gym.envs.observation import ObservationType
from f1tenth_gym.envs.reset import ResetStrategy
from f1tenth_gym.envs.env_config import EnvConfig, ObservationConfig, ResetConfig

work = {
    "mass": 3.463388126201571,
    "lf": 0.15597534362552312,
    "tlad": 0.82461887897713965,
    "vgain": 1,
}

cfg = EnvConfig(
    observation_config=ObservationConfig(type=ObservationType.KINEMATIC_STATE),
    reset_config=ResetConfig(strategy=ResetStrategy.RL_RANDOM_STATIC),
)

env = gym.make(
    "f1tenth_gym:f1tenth-v0",
    config=cfg,
    render_mode="rgb_array",
)
env = gymnasium.wrappers.RecordVideo(env, f"video_{time.time()}")
track = env.unwrapped.track

track.raceline.render_waypoints(env.unwrapped.renderer)
poses = np.array(
    [
        [
            track.raceline.xs[0],
            track.raceline.ys[0],
            track.raceline.yaws[0],
        ]
    ]
)

obs, info = env.reset(options={"poses": poses})
done = False

laptime = 0.0
start = time.time()

frames = [env.render()]
while not done and laptime < 5.0:
    action = env.action_space.sample()
    for i, agent_id in enumerate(obs.keys()):
        speed, steer = 1.0, 0.2
        action[i] = [steer, speed]

    obs, step_reward, done, truncated, info = env.step(action)
    laptime += step_reward

    frame = env.render()
    frames.append(frame)

print("Sim elapsed time:", laptime, "Real elapsed time:", time.time() - start)

# close env to trigger video saving
env.close()

```
